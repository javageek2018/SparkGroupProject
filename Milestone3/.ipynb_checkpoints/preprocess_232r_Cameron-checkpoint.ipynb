## Imputing

Missing values were concentrated in policy-related fields such as airport_fee and congestion_surcharge.

For this dataset, a missing airport fee logically means the trip was not an airport trip, so it is reasonable to treat missing values as zero.

Similarly, missing congestion surcharge values likely indicate that the surcharge did not apply to that trip.

Rather than dropping millions of rows, we impute these values to preserve data volume and avoid introducing bias.

```python
df = df.fillna({
    "airport_fee": 0,
    "congestion_surcharge": 0
})

df = df.withColumn(
    "wav_match_flag",
    F.when(
        F.col("wav_match_flag").isNull() |
        (F.length(F.trim(F.col("wav_match_flag").cast("string"))) == 0),
        F.lit("Unknown")
    ).otherwise(F.trim(F.col("wav_match_flag").cast("string")))
)

df.groupBy("wav_match_flag").count().show()
```

```python
null_summary = df.select([
    F.count(F.when(F.col(c).isNull(), c)).alias(c)
    for c in df.columns
])

null_summary.toPandas().T
```

## Scaling/Transforming

```python
num_cols = [
    "log_fare",
    "log_miles",
    "log_minutes",
    "tolls",
    "congestion_surcharge",
    "airport_fee",
    "pickup_hour",
    "pickup_dow",
    "pickup_month",
    "is_weekend",
    "is_night",
    "has_airport_fee",
    "has_congestion_fee",
    "total_surcharges",
    "fare_per_mile"
]

# target (tip)
df1 = df.withColumn("tip_binary", F.when(F.col("tips").cast("double") > 0, 1).otherwise(0))

# numeric sanity casts first
df1 = (
    df1
    .withColumn("base_passenger_fare", F.col("base_passenger_fare").cast("double"))
    .withColumn("trip_miles", F.col("trip_miles").cast("double"))
    .withColumn("trip_time", F.col("trip_time").cast("double"))
    .withColumn("tolls", F.col("tolls").cast("double"))
    .withColumn("bcf", F.col("bcf").cast("double"))
    .withColumn("sales_tax", F.col("sales_tax").cast("double"))
    .withColumn("congestion_surcharge", F.col("congestion_surcharge").cast("double"))
    .withColumn("airport_fee", F.col("airport_fee").cast("double"))
    .withColumn("tips", F.col("tips").cast("double"))
)

# time features
df1 = (
    df1
    .withColumn("trip_minutes", F.col("trip_time") / F.lit(60.0))
    .withColumn("pickup_hour", F.hour("pickup_datetime").cast("double"))
    .withColumn("pickup_dow", F.dayofweek("pickup_datetime").cast("double"))
    .withColumn("pickup_month", F.month("pickup_datetime").cast("double"))
    .withColumn("log_fare", F.when(F.col("base_passenger_fare") >= 0, F.log1p("base_passenger_fare")).otherwise(None))
    .withColumn("log_miles", F.when(F.col("trip_miles") >= 0, F.log1p("trip_miles")).otherwise(None))
    .withColumn("log_minutes", F.when(F.col("trip_minutes") >= 0, F.log1p("trip_minutes")).otherwise(None))
)

# flags
df1 = (
    df1
    .withColumn("is_weekend", F.when(F.dayofweek("pickup_datetime").isin([1, 7]), 1.0).otherwise(0.0))
    .withColumn("is_night", F.when(F.hour("pickup_datetime").isin(list(range(0,6)) + list(range(20,24))), 1.0).otherwise(0.0))
    .withColumn("has_airport_fee", F.when(F.col("airport_fee") > 0, 1.0).otherwise(0.0))
    .withColumn("has_congestion_fee", F.when(F.col("congestion_surcharge") > 0, 1.0).otherwise(0.0))
)

df1 = (
    df1
    .withColumn(
        "total_surcharges",
        F.coalesce("tolls", F.lit(0.0)) +
        F.coalesce("bcf", F.lit(0.0)) +
        F.coalesce("sales_tax", F.lit(0.0)) +
        F.coalesce("congestion_surcharge", F.lit(0.0)) +
        F.coalesce("airport_fee", F.lit(0.0))
    )
)

df1 = (
    df1
    .withColumn(
        "fare_per_mile",
        F.when(F.col("trip_miles") > 0, F.col("base_passenger_fare") / F.col("trip_miles")).otherwise(None)
    )
)

df1.select(
    "tip_binary", "log_fare", "log_miles", "log_minutes", "pickup_hour", 
    "pickup_dow", "pickup_month","is_weekend","is_night",
    "has_airport_fee","has_congestion_fee",
    "total_surcharges", "fare_per_mile"
).limit(5).toPandas()
```

```python
# 1. Convert NaNs to Nulls without changing column names
df_num = df1.replace(float('nan'), None, subset=num_cols)

# 2. Imputation
imputer = Imputer(
    inputCols=num_cols,
    outputCols=[f"{c}_imp" for c in num_cols],
    strategy="median"
)
imputer_model = imputer.fit(df_num)
df_imp = imputer_model.transform(df_num)

# 3. Assemble
assembler_num = VectorAssembler(
    inputCols=[f"{c}_imp" for c in num_cols],
    outputCol="num_features",
    handleInvalid="error"
)
df2 = assembler_num.transform(df_imp)

# 4. Scale: Standardize the features
scaler = StandardScaler(
    inputCol="num_features", 
    outputCol="num_scaled", 
    withMean=True, 
    withStd=True
)
scaler_model = scaler.fit(df2)
df3 = scaler_model.transform(df2)

# 5. View
df3.select("num_features", "num_scaled", "tip_binary").limit(5).toPandas()
```

```python
print("Rows:", df3.count())
df3.groupBy("tip_binary").count().show()
```

## Encoding

```python
# Categorical

cat_cols = [
    "hvfhs_license_num",
    "shared_request_flag",
    "shared_match_flag",
    "access_a_ride_flag",
    "wav_request_flag",
    "wav_match_flag"
]

df5 = df3

for c in cat_cols:
    df5 = df5.withColumn(c, F.trim(F.col(c).cast("string")))
    df5 = df5.withColumn(
        c,
        F.when(F.col(c).isNull() | (F.length(F.col(c)) == 0), F.lit("Unknown")).otherwise(F.col(c))
    )

for c in cat_cols:
    print(c)
    pdf = df5.groupBy(c).count().orderBy(F.desc("count")).limit(10).toPandas()
    display(pdf)
```

```python
# StringIndex + OneHot Encode

indexers = [
    StringIndexer(inputCol=c, outputCol=f"{c}_idx", handleInvalid="keep")
    for c in cat_cols
]

encoder = OneHotEncoder(
    inputCols=[f"{c}_idx" for c in cat_cols],
    outputCols=[f"{c}_ohe" for c in cat_cols],
    handleInvalid="keep"
)

df_idx = df5
for idx in indexers:
    df_idx = idx.fit(df_idx).transform(df_idx)

df_enc = encoder.fit(df_idx).transform(df_idx)
```

```python
# Check new dataset
df_enc.select(
    cat_cols +
    [f"{c}_idx" for c in cat_cols] +
    [f"{c}_ohe" for c in cat_cols]
).limit(3).toPandas().T
```

## Feature Engineering

```python
assembler_all = VectorAssembler(
    inputCols=["num_scaled"] + [f"{c}_ohe" for c in cat_cols],
    outputCol="features",
    handleInvalid="keep"
)

df_final = assembler_all.transform(df_enc).select(
    F.col("tip_binary").cast("int").alias("label"),
    "features",
    "pickup_datetime",
    "PULocationID"
)

df_final.limit(5).toPandas()
```

```python
df_final.write.mode("overwrite").parquet("processed_data/vectorized_features")
```

# Modeling

```python
df_final = spark.read.parquet("processed_data/vectorized_features")
```

```python
df_model = df_final.select("label", "features")
```

```python
# Verify class imbalance
df_model = df_final.select("label", "features")
df_model.groupBy("label").count().show()
```

```python
# Stratified 30% subsample
df_model = df_model.sampleBy("label", fractions={0: 0.3, 1: 0.3}, seed=42)
```

```python
# Verify class imbalance of subsample
df_model.groupBy("label").count().show()
```

```python
# Calculate balancing ratio
total = 628436051 + 116837270
pos   = 116837270
balancing_ratio = (total - pos) / pos  # ~5.38

# Add class weight column
df_model = df_model.withColumn(
    "class_weight",
    when(col("label") == 1, balancing_ratio).otherwise(1.0)
)

# Split
train_df, test_df = df_model.randomSplit([0.8, 0.2], seed=42)
```

```python
print("Train size:", train_df.count())
print("Test size:", test_df.count())
train_df.groupBy("label").count().show()
test_df.groupBy("label").count().show()
```

## Random Forest Classifier

```python
rf = RandomForestClassifier(
    labelCol="label",
    featuresCol="features",
    weightCol="class_weight",
    numTrees=20,
    maxDepth=10,
    maxBins=16,
    subsamplingRate=0.7,
    featureSubsetStrategy="sqrt",
    seed=42
)

rf_model1 = rf.fit(train_df)
```

# Preprocessing Summary

The following preprocessing steps were completed using Spark DataFrame 
operations and Spark MLlib transformers to prepare the raw FHVHV ride-share 
dataset for distributed model training.

## 1. Imputing

Missing values were concentrated in two policy-related numeric fields. 
Rather than dropping millions of affected rows, domain-informed imputation 
was applied to preserve data volume:

- **airport_fee:** Null values replaced with 0.0. A missing airport fee 
  logically indicates the trip did not originate from an airport, making 
  zero the semantically correct fill value.
- **congestion_surcharge:** Null values replaced with 0.0 for the same 
  reasoning — absence of the surcharge indicates the trip did not pass 
  through a congestion pricing zone.
- **wav_match_flag:** Null and empty string values replaced with "Unknown" 
  to ensure the categorical encoder could handle all records without 
  dropping rows.

For the remaining numeric features (log_fare, log_miles, log_minutes, 
tolls, and engineered features), Spark MLlib's `Imputer` was applied using 
a **median strategy** to handle any residual nulls introduced during 
feature engineering without distorting the distributions of skewed 
financial variables.

## 2. Scaling and Transforming

Raw numeric features were transformed to reduce skewness and normalize 
scale before model training:

- **Log transforms:** `base_passenger_fare`, `trip_miles`, and 
  `trip_minutes` were log1p transformed (log(1 + x)) to compress the 
  heavy right-skewed distributions observed in the EDA. This prevents 
  large fare and distance outliers from dominating the feature space.
- **StandardScaler:** All numeric features were standardized using Spark 
  MLlib's `StandardScaler` with `withMean=True` and `withStd=True`, 
  producing zero-mean, unit-variance features. This ensures that features 
  on different scales (e.g. pickup_hour ranging 0-23 vs. log_fare ranging 
  0-5) contribute proportionally to model training.


## 3. Encoding

Six categorical variables were encoded for use in the model:

- **hvfhs_license_num** — service provider (Uber, Lyft, etc.)
- **shared_request_flag** — whether a shared ride was requested
- **shared_match_flag** — whether the shared ride was matched
- **access_a_ride_flag** — accessibility ride indicator
- **wav_request_flag** — wheelchair-accessible vehicle request
- **wav_match_flag** — wheelchair-accessible vehicle match

Each variable was first passed through `StringIndexer` to convert string 
labels to numeric indices, then through `OneHotEncoder` to produce sparse 
binary vectors. The `handleInvalid="keep"` parameter was set on both 
transformers to ensure unseen or null categories do not cause runtime 
failures.

## 4. Feature Engineering

Several new features were constructed using Spark SQL functions to capture 
temporal patterns and trip economics that are not directly represented in 
the raw columns:

- **pickup_hour** — hour of day (0-23), captures time-of-day tipping patterns
- **pickup_dow** — day of week (1-7), captures weekday vs. weekend behavior
- **pickup_month** — month of year, captures seasonal trends
- **is_weekend** — binary flag for Saturday/Sunday trips
- **is_night** — binary flag for trips between 8pm and 6am
- **has_airport_fee** — binary flag indicating airport pickup
- **has_congestion_fee** — binary flag indicating congestion zone trip
- **total_surcharges** — sum of all surcharge components (tolls, bcf, 
  sales_tax, congestion_surcharge, airport_fee)
- **fare_per_mile** — base_passenger_fare divided by trip_miles, 
  capturing trip cost efficiency

## 5. Final Feature Assembly

All preprocessed numeric and encoded categorical features were combined 
into a single sparse feature vector using `VectorAssembler`, producing a 
39-dimensional feature vector per row. The final dataset was written to 
Parquet at `processed_data/vectorized_features` with the schema:

- **label** — binary tip indicator (1 if tips > 0, else 0)
- **features** — 39-dimensional sparse vector (15 scaled numeric + 
  24 OHE categorical dimensions)

The full preprocessing pipeline is reproducible: reading from the 
deduplicated Parquet dataset and executing all cells in order will 
reproduce the exact feature matrix used for model training. When 
paired with the EDA notebook an identical deduplicated dataset can
also be reproduced.

# Fitting Analysis

## Where does the model fit in the fitting graph?

Both Random Forest models sit firmly in the **underfitting region** of the 
bias-variance tradeoff. The train/test gaps were essentially zero across 
both configurations (gap of -0.000024 for RF1 and 0.000363 for RF2), 
meaning the models generalize exceedingly well to unseen data but are not complex 
enough to fully capture the predictive signal in the dataset. On the 
classical fitting curve, both models sit to the left of the optimal 
complexity point — increasing model capacity is expected to improve both 
train and test performance simultaneously before overfitting becomes a 
concern.

### Hyperparameter Comparison

| Model | numTrees | maxDepth | maxBins | Train ROC-AUC | Test ROC-AUC | Train PR-AUC | Test PR-AUC | Gap     |
|-------|----------|----------|---------|---------------|--------------|--------------|-------------|---------|
| RF1   | 20       | 10       | 16      | 0.6443        | 0.6443       | 0.2427       | 0.2425      | ~0.000  |
| RF2   | 10       | 15       | 12      | 0.6530        | 0.6526       | 0.2481       | 0.2476      | 0.0004  |

Increasing maxDepth from 10 to 15 while reducing numTrees from 20 to 10 
produced a consistent ~0.008 improvement in ROC-AUC and ~0.005 improvement 
in PR-AUC. This indicated that depth was a crucial hyperparameter for this 
dataset — deeper trees are better able to capture the complex feature 
interactions that drive tipping behavior. 

## Which model performs best and why?

**RF Model 2 performs best** across all four metrics. The improvement is 
modest but consistent — both ROC-AUC and PR-AUC moved in the right 
direction with no increase in overfitting. The deeper tree structure 
allows the model to learn multi-condition rules that shallower 
trees cannot represent. The fact that the train/test gap remained near 
zero despite increased depth indicates that the model has not yet reached 
its capacity ceiling and that further depth increases might continue to 
improve performance.

## Next Models for Milestone 4

**Gradient Boosted Trees (GBT):** GBT is the most natural next step given 
the current results. Unlike Random Forest which builds trees independently 
in parallel, GBT builds trees sequentially where each tree corrects the 
errors of the previous one. This sequential error-correction mechanism is 
probably better suited to the noisy, imbalanced nature of tipping prediction.
We have already defined two GBT configurations in this notebook but have yet
to successfully run them on the 30% stratified subsample of our data.

**XGBoost (Distributed):** XGBoost with the scale_pos_weight parameter 
for handling the class imbalance is one that our group feels could provide 
a significant boost in performance due to its more robust regularization and 
optimization.

# Conclusion

## What is the conclusion of the first model?

Two Random Forest classifiers were trained on a 30% stratified subsample 
of the FHVHV ride-share dataset (~223M rows) to predict tipping behavior 
(binary: tip > 0 vs. no tip). Both models achieved a test ROC-AUC of 
approximately 0.644-0.653, meaningfully above the 0.50 random baseline 
established by the initial unweighted model when we first tested. 
The near-zero train/test gap across both configurations confirms that the 
models generalize well but are likely underfitting due to not having learned
enough complexity to overfit, but are also probably leaving predictive signal on the table.

The most important finding from hyperparameter comparison is that depth 
is the primary driver, or at least a very significant driver, of performance 
on this dataset. Increasing maxDepth from 10 to 15 (while reducing numTrees 
from 20 to 10) produced a consistent ~0.008 improvement in ROC-AUC, suggesting 
that tipping behavior is driven by complex feature interactions such as 
combinations of fare size, time of day, and trip type rather than simple 
thresholds that shallow trees can capture.

Class imbalance (84% no-tip vs. 16% tip) was addressed through inverse 
frequency class weighting (~5.38x weight on positive class).

## What can be done to possibly improve it?

Several avenues exist for improving model performance. I'll discuss a few ideas below:

**Model capacity:** One avenue of improvement lies in continuing to tune hyperparameters.
We could increase the number of trees, the number of bins, or perhaps use more of our data
than the current 30% stratified subsample. Doing any of these could require modifying the
curent spark configuration to increase driver memory or overall memory per core.

**Feature engineering:** The current feature set relies primarily on 
trip-level attributes (fare, distance, time of day). Aggregated historical 
features — such as average tip rate by pickup zone, by hour of day, or by 
service provider (hvfhs_license_num) — could add substantial 
predictive signal that trip-level features alone cannot capture.

**Geographic features:** PULocationID and DOLocationID were excluded from 
the current models due to high cardinality. We could potentionally encode
these another way (e.g., mean tip rate per zone) rather than OHE. This might
allow the model to leverage geographic signal without the memory cost of
high-dimensional sparse vectors.

**Alternative models:** Gradient Boosted Trees (GBT) or other models might
outperform Random Forest on this problem due to the noisy, imbalanced 
nature of tipping data. 

## How did distributed computing help with this task?

Training a binary classifier on 745M ride-share records — even subsampled 
to 223M rows — would be completely infeasible on a single-threaded process. 
The Spark distributed computing framework provided several critical advantages:

**Parallel data processing:** The feature engineering pipeline was executed in 
parallel across all available cores, reducing preprocessing time from what would 
be hours on a single thread to a more manageable timeframe.

**Distributed model training:** Spark MLlib's RandomForestClassifier 
distributes the tree-building process across workers, with each worker 
computing splits on its local data partition. This allowed 
training on tens of millions of rows per tree without requiring all data 
to fit in a single machine's memory.

**Scalable shuffle and aggregation:** Operations like dropDuplicates, 
groupBy, and sampleBy on 745M rows rely on Spark's distributed shuffle 
mechanism to move and aggregate data across partitions — operations that 
pandas or scikit-learn could not perform at this scale. Even with this
capability we OOM'd several times figuring out how to work through these
intensive operations.

**Out-of-core processing:** Spark's ability to spill intermediate results 
to disk (via spark.local.dir on the scratch filesystem) allowed the 
pipeline to process data volumes that exceed available RAM, which was 
critical for the deduplication and preprocessing stages.
